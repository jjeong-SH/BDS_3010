```python
import pandas as pd
import matplotlib.pyplot as plt
import os
os.chdir('./data')
```

```python
data = pd.read_csv("./Samsung_BioLogics.csv",encoding = 'cp949',engine='python')
datasort = data.sort_values(by='')
datasort.set_index('',inplace=True)
D1 = datasort[''].loc['2020/03/03':'2020/03/10']
D2 = datasort[''].loc['2020/08/14':'2020/09/02']
D3 = datasort[''].loc['2020/12/15':'2021/01/10']
s1 = datasort['']
plt.figure(figsize=(15,7))
ax1 = s1.plot()                              # :datasort[datasort['']==] 
plt.axvspan(287,292,color = 'lightcoral')          #1 
plt.axvspan(401,413,color = 'lightcoral')          #2 
plt.axvspan(483,499,color = 'lightcoral')          #3 

D1 = D1/D1['2020/03/03']
D2 = D2/D2['2020/08/14']
D3 = D3/D3['2020/12/15']
plt.xlabel('Date')
plt.ylabel('Price(KRW)')
plt.title('Samsung_BioLogics')
plt.figure(figsize=(10,5))
D1.plot()
plt.xlabel('Date')
plt.ylabel('Price rate')
plt.title('D1')
plt.figure(figsize=(10,5))
D2.plot()
plt.xlabel('Date')
plt.ylabel('Price rate')
plt.title('D2')
plt.figure(figsize=(10,5))
D3.plot()
plt.xlabel('Date')
plt.ylabel('Price rate')
plt.title('D3')
plt.show()
```
